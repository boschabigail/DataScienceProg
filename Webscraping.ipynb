```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
import seaborn as sns
```

This statement creates the connection between the html page and the notebook

```python
url='https://www.boxofficemojo.com/year/2022/?ref_=bo_lnav_hm_shrt'
response = requests.get(url)
status = response.status_code
if status == 200:
    page = response.text
    soup = bs(page)
else:
    print(f"Oops! Recieved Status code {status}")
```

```python
print(soup)
```

The code below circles through the table to find the name of every movie in the table.
And the one below that circles through for the release date

```python
movie_title = []
table = soup.find_all('td',class_='a-text-left mojo-field-type-release mojo-cell-wide')
table[0].select('a')[0].string
```

```python
for row in table:
    movie_title.append(row.select('a')[0].string)
movie_title
```

```python
release = []
release_date=soup.find_all('td',class_ ='a-text-left mojo-field-type-date a-nowrap')
release_date[0].string
for row in release_date:
    release.append(row.string)
release
```

```python
total_gross = []
gross=soup.find_all('td',class_='a-text-right mojo-field-type-money mojo-estimatable')
for row in gross:
    total_gross.append(row.string)
total_gross
gross = []
for x in total_gross:
    if x not in gross:
        gross.append(x)
gross
```

```python
theater_amt = []
theater=soup.find_all('td',class_='a-text-right mojo-field-type-positive_integer')
for row in theater:
    theater_amt.append(row.string)
theater_amt
```

The above code shows the amount of theaters the movie was released in.
The code below shows the distributors of each movie released

```python
distributors = []
distrib = soup.find_all('td',class_='a-text-left mojo-field-type-studio')
distrib[0].select('a')
for row in distrib:
    distributors.append(row.select('a'))
distributors
```

```python
test_df = []
length = len(movie_title)
i = 0
while i < length:
    test_df.append({'title':movie_title[i],'release date':release[i],'theaters':theater_amt[i],'gross':gross[i],'distributors':distributors[i]})
    i += 1
test_df
```

The above code sorts all of the entries into a list where they are connected to the appropriate values
Then that list was made into a data frame that could be manipulated further.

```python
len(movie_title)
```

```python
movie_df = pd.DataFrame(test_df)
movie_df
```
